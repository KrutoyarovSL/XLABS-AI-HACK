Описание решения:

---



Для воспроизведения вокала мы дообучили модель stabilityai/stable-audio-open-1.0 (https://huggingface.co/stabilityai/stable-audio-open-1.0) до генерации вокала на русском языке по текстовому запросу. Текстовый запрос должен представлять из себя готовый текст песни для его последующего воспроизведения. Пользователь обращается к нашей модели через чат-бота в телеграмме. После ввода команды /start у пользователя будет выбор: использовать свой текст для генерации вокала/сгенерировать текст с помощью GigaChat (это наша фича). После получения текста от GigaChat или ввода своего текста запрос поступает в модель и на "выходе" пользователь получает mp3 файл.
